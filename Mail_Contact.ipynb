```python
import os
import pandas as pd
from datetime import datetime  
import smtplib
from email.message import EmailMessage
```

```python
# ^^^^^^^^^^^^^^^^  LOAD EXCEL FILE ***********************
```

```python
mail_contact = 'Company_Contact.xlsx'
```

```python
# ----------------- Read Excel and fix dtypes-----------------------
```

```python
if os.path.exists(mail_contact):
    df = pd.read_excel(mail_contact)
    print("✔️ File loaded successfully.")
else:
    print(f"💀 File '{mail_contact}' is missing in the path or repository.")
```

```python
#  ----------------- Fix data types -----------------------------
```

```python
df['Status'] = df['Status'].astype(str)
df['Time'] = pd.to_datetime(df['Time'], errors='coerce') 
```

```python
# --------------------  CONNECT TO GMAIL SMTP --------------------
```

```python
server = smtplib.SMTP('smtp.gmail.com', 587)
server.starttls()
server.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
```

```python
# --------------------  Mail Contact  --------------------
```

```python
for i in range(len(df)):
    try:
        status = str(df.at[i, 'Status']).strip().lower() if pd.notna(df.at[i, 'Status']) else ''
        if status == 'sent':
            print(f"⏭️ Row {i+1}: Already sent")
        else:
            to_email = df.at[i, 'Company_Mail_id']
            company = df.at[i, 'Company_Name']
            Role = df.at[i, 'Role']
            attachment_path = str(df.at[i, 'Attachment']).strip()

            if os.path.exists(attachment_path):
                msg = EmailMessage()
                msg['Subject'] = f'Application to {company} - {Role}'
                msg['From'] = EMAIL_ADDRESS
                msg['To'] = to_email

                hr_name = df.at[i, 'HR_Firstname']
                greeting = f"Dear {hr_name}, "

                msg.set_content(f"""{greeting}

I hope this message finds you well.

I would like to express my keen interest in {Role} opportunity in {company}.

Currently working as a functional business analyst in immigration domain. However, I have strong knowledge in the domain of HR, finance, immigration and healthcare. Additionally, I have 6 plus years of experience in HR Reporting & Workforce analytics roles.

Please find my resume attached for your consideration. I would be grateful for the opportunity to discuss how I can add value to your team.

Looking forward to hearing from you.

Regards,
Mridhul TP
8089778087
Kochi, Kerala
""")


                with open(attachment_path, 'rb') as f:
                    file_data = f.read()
                    file_name = os.path.basename(attachment_path)
                    msg.add_attachment(file_data, maintype='application', subtype='octet-stream', filename=file_name)

                server.send_message(msg)
                print(f"✔️ Row {i+1}: Email sent to {to_email}")

                #  Update status and time with correct types
                df.at[i, 'Status'] = 'Sent'
                df.at[i, 'Time'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

            else:
                print(f"❌ Row {i+1}: Attachment not found: {attachment_path}")

    except Exception as e:
        print(f"❌ Row {i+1}: Error: {e}")

# 💾 Save updated Excel
df.to_excel(mail_contact, index=False)
server.quit()
print("Done.")
```
